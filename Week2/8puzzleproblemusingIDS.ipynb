```python
from copy import deepcopy

goal_state = [[1, 2, 3],
              [4, 5, 6],
              [7, 8, 0]]

def is_goal(state):
    return state == goal_state

def find_zero(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j

def get_neighbors(state):
    neighbors = []
    x, y = find_zero(state)
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]  # Up, Down, Left, Right

    for dx, dy in directions:
        nx, ny = x + dx, y + dy
        if 0 <= nx < 3 and 0 <= ny < 3:
            new_state = deepcopy(state)
            new_state[x][y], new_state[nx][ny] = new_state[nx][ny], new_state[x][y]
            neighbors.append(new_state)
    return neighbors

def DLS(state, depth, max_depth, visited, path):
    if is_goal(state):
        return path + [state]
    if depth > max_depth:
        return None

    visited.add(str(state))  # Use string representation instead of tuple

    for neighbor in get_neighbors(state):
        if str(neighbor) not in visited:
            result = DLS(neighbor, depth + 1, max_depth, visited.copy(), path + [state])
            if result is not None:
                return result
    return None

def IDS(initial_state, max_depth_limit=50):
    for depth_limit in range(max_depth_limit + 1):
        print(f"Searching at depth limit: {depth_limit}")
        visited = set()
        result = DLS(initial_state, 0, depth_limit, visited, [])
        if result is not None:
            return result
    return None

initial_state = [[0,1,2],
                 [4,5,3],
                 [7,8,6]]

solution = IDS(initial_state)

if solution:
    print(f"\nSolution found in {len(solution) - 1} moves:\n")
    for step in solution:
        for row in step:
            print(row)
        print()
else:
    print(" No solution found within depth limit.")
```
