```python
from copy import deepcopy

goal_state = [[1, 2, 3],
              [4, 5, 6],
              [7, 8, 0]]

def is_goal(state):
    return state == goal_state

def find_zero(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j

def get_neighbors(state):
    neighbors = []
    x, y = find_zero(state)
    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]  # Up, Down, Left, Right

    for dx, dy in directions:
        nx, ny = x + dx, y + dy
        if 0 <= nx < 3 and 0 <= ny < 3:
            new_state = deepcopy(state)
            # Swap 0 with the target tile
            new_state[x][y], new_state[nx][ny] = new_state[nx][ny], new_state[x][y]
            neighbors.append(new_state)
    return neighbors

# Recursive DFS function
def DFS(state, visited, path, max_depth=50):
    if is_goal(state):
        return path + [state]

    if len(path) > max_depth:
        return None

    visited.add(str(state))

    for neighbor in get_neighbors(state):
        neighbor_str = str(neighbor)
        if neighbor_str not in visited:
            result = DFS(neighbor, visited.copy(), path + [state], max_depth)
            if result is not None:
                return result
    return None

def solve_dfs(initial_state, max_depth=50):
    visited = set()
    return DFS(initial_state, visited, [], max_depth)

initial_state = [[1, 2, 0],
                 [4, 5, 3],
                 [7, 8, 6]]

solution = solve_dfs(initial_state)

if solution:
    print(f"\nSolution found in {len(solution) - 1} moves:\n")
    for step in solution:
        for row in step:
            print(row)
        print()
else:
    print("No solution found within depth limit.")
```
