```python
def vacuum_world():
    goal = {'A': '0', 'B': '0'}
    cost = 0

    loc = input("Enter Vacuum Location (A or B): ").upper()
    stat_cur = input(f"Enter status of Room {loc} (0=Clean, 1=Dirty): ")
    stat_other = input("Enter status of the other room (0=Clean, 1=Dirty): ")

    print("\nInitial State:")
    print(f"Vacuum in: {loc}")
    print(f"Room {loc}: {'Dirty' if stat_cur == '1' else 'Clean'}")
    print(f"Other Room: {'Dirty' if stat_other == '1' else 'Clean'}")

    if loc == 'A':
        print("\nVacuum at Room A")
        if stat_cur == '1':
            print("Cleaning A")
            goal['A'] = '0'
            cost += 1
        if stat_other == '1':
            print("Moving to B")
            cost += 1
            print("Cleaning B")
            goal['B'] = '0'
            cost += 1
        else:
            print("Room B already clean")

    elif loc == 'B':
        print("\nVacuum at Room B")
        if stat_cur == '1':
            print("Cleaning B")
            goal['B'] = '0'
            cost += 1
        if stat_other == '1':
            print("Moving to A")
            cost += 1
            print("Cleaning A")
            goal['A'] = '0'
            cost += 1
        else:
            print("Room A already clean")
    else:
        print("Invalid Location")
        return

    print("\nGOAL STATE:")
    print(goal)
    print("Total Cost:", cost)

vacuum_world()
```
